```python
import gym
import numpy as np
from tqdm import tqdm
from matplotlib import pyplot as plt
from Cartpole.agents import QLearningAgent
```

```python
# Initialize learning environment and agent
env = gym.make('CartPole-v0')
```

```python
# Create agent
agent = QLearningAgent(env)

# Play N episodes to train the agent
N = 1000
total_rewards = np.empty(N)
for n in tqdm(range(N)):
    eps = 1.0/np.sqrt(n+1)
    total_reward = agent.play(epsilon=eps, gamma=0.99)
    total_rewards[n] = total_reward
```

```python
def plot_running_avg(totalrewards):
    N = len(totalrewards)
    running_avg = np.empty(N)
    for t in range(N):
        running_avg[t] = totalrewards[max(0, t-100):(t+1)].mean()
    plt.plot(running_avg)
    plt.title("Running Average")
    plt.show()
```

```python
# Sanity check on training
plt.plot(total_rewards)
plt.title("Rewards")
plt.show()
```

```python
plot_running_avg(total_rewards)
```

```python
# Test if the agent is properly trained
agent.play(epsilon=0, render=True)
```

```python
# Reset environment and set the learning agent to use Scikit-learn
env.reset()
agent2 = QLearningAgent(env, use_sklearn=True)
```

```python
# Play N episodes to train the agent
N = 1000
total_rewards = np.empty(N)
for n in tqdm(range(N)):
    eps = 1.0/np.sqrt(n+1)
    total_reward = agent2.play(epsilon=eps, gamma=0.99)
    total_rewards[n] = total_reward
```

```python
plot_running_avg(total_rewards)
```

```python
# Test if the agent is properly trained
agent2.play(epsilon=0, render=True)
```
